
### Jupyter Installation and Plumbing

| Package | Role in the System | Why You Need It |
|:---|:---|:---|
| **`jupyterlab`** (or `notebook`) | **The Server/Interface:** This is the application that launches the web browser interface, manages files, and hosts the entire notebook environment. | You need this to **start** Jupyter (e.g., `$ jupyter lab`). It's the front end. |
| **`ipykernel`** | **The Connector/Bridge:** This package tells the `jupyterlab` server *where* to find and *how* to talk to the specific Python executable inside your `.venv`. | You need this to register your custom **`uv`** environment as an available **Kernel** for running code. |
| **`yfinance`**, etc. | **The Libraries:** The actual packages that execute your data analysis code. | You need this to run your cell code without a `ModuleNotFoundError`. |
